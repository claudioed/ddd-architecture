```python
ddd_tactical = {
    "guides_4_microservices" : {
        "srp" : "Single Responsibility Principle (SRP): Apply SRP to microservices. A microservice should do one thing well and have a single reason to change.",
        "manageable" : "Keep microservices small and manageable. However, small is relative and context-dependent; focus on cohesion and autonomy rather than size." ,
        "ddd" : "DDD principles to identify bounded contexts. Each bounded context corresponds to a specific domain model and has explicit boundaries, making it a candidate for a separate microservice."
    }
}
```

### Describe the domain

Here we will describe the domain that we will explore


```python
domain = """
  The Payment Gateway domain encompasses the infrastructure and services that facilitate online payment processing between merchants and customers. It acts as an intermediary, securely transmitting payment information from the customer to the merchant's bank account or payment processor. This domain includes handling authorization of transactions, encryption of payment details, fraud detection, and compliance with financial standards and regulations. Payment gateways are essential for e-commerce, enabling businesses to accept various forms of digital payments (such as credit cards, bank transfers, and digital wallets) while ensuring transaction security and integrity.
"""
```

### Describe the subdomain

Here we will describe the subdomain that we want some help

```python
subdomain = """Fraud Detection and Risk Management this subdomain is responsible for identifying and mitigating fraudulent activities. It employs various tools and techniques, such as machine learning algorithms, behavior analysis, and risk scoring systems, to detect suspicious transactions and prevent fraud."""
```

### Define the Bounded Context 

Here we will describe the bounded context that we will work

```python
bounded_context = "This bounded context is dedicated to the detection and prevention of fraudulent activities. It uses machine learning, behavior analysis, and risk scoring to identify and block suspicious transactions, thereby reducing the risk of fraud."
```

### AI to help uo in our services

Here we will discover the suggested services for our Bounded Context

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage,AIMessage
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains import SequentialChain

model = ChatOpenAI(
    model="gpt-4o",
    temperature=0,
)

messages = [
    SystemMessage(content="You are a solution architect. Your main assignment is create software documentation"),
    AIMessage(content="I want you to act as an IT Architect. I will provide some details about the functionality of an application or other digital product, and it will be your job to come up with ways to integrate it into the IT landscape. This could involve analyzing business requirements, performing a gap analysis and mapping the functionality of the new system to the existing IT landscape. Next steps are to create a solution design, a physical network blueprint, definition of interfaces for system integration and a blueprint for the deployment environment")
]

bounded_context_prompt = PromptTemplate(
    input_variables=["bounded_context","domain","subdomain","guides_4_microservices"],
    template="""

        Your task is to perform the following actions:
        1. Analyze the domain, subdomain, and bounded context as input
        2. Define the microservices for {bounded_context} using domain and subdomain as a context  
        3. Identify the microservice following the guides.
        4. Give me the microservices list and description of each one. 
        5. We don't need any source code. This is a documentation stuff.

        Domain: {domain}
        Subdomain: {subdomain}
        Bounded context: {bounded_context}
        
        Guides: {guides_4_microservices}
        
    """
)

ddd_chain = LLMChain(llm=model,prompt=bounded_context_prompt) 
final_chain = SequentialChain(chains=[ddd_chain],input_variables=["bounded_context","domain","subdomain","guides_4_microservices"],verbose=True)
response = final_chain.invoke({'bounded_context': bounded_context, 'domain':domain, 'subdomain':subdomain,'guides_4_microservices' : ddd_tactical['guides_4_microservices']})

print(response['text'])
```

```python
detailed_service = """
    Transaction Monitoring Service
    Description: Provides real-time monitoring of transactions to detect suspicious activities. Generates alerts and logs for further investigation by security teams.
"""

service_detail_prompt = PromptTemplate(
    input_variables=["detailed_service"],
    template="""

        Your task is to analyse the bounded context in detail and generate the following :
        
        1. Using Domain Driven Design techniques, Identify the Entities, Value Objects with list of attributes and behaviours for each.
        2. List the services, repositories, etc. required for the application. 
        3. We don't need any source code. This is a documentation stuff.
        4. Do it step-by-step and give me the details of each step.

        Bounded Context: {detailed_service}
        
    """
)

service_chain = LLMChain(llm=model,prompt=service_detail_prompt) 

final_chain = SequentialChain(chains=[service_chain],input_variables=["detailed_service"],verbose=True)
response = final_chain.invoke({'detailed_service': detailed_service})
service_elements = response['text']
print(response['text'])
```

```python
service_class_diagram_prompt = PromptTemplate(
    input_variables=["service_elements"],
    template="""

        Your task is to analyse Entities, Value Objects in detail and generate the following :
        1. Generate the class diagram for the Entities and Value Objects, there is no need for Repository and Services
        2. Give me diagram using mermaid as syntax in .md format 

        Service Context: {service_elements}
        
    """
)

service_class_chain = LLMChain(llm=model,prompt=service_class_diagram_prompt) 
final_chain = SequentialChain(chains=[service_class_chain],input_variables=["service_elements"],verbose=True)
response = final_chain.invoke({'service_elements': service_elements})

print(response['text'])
```

```python
service_class_diagram_prompt = PromptTemplate(
    input_variables=["service_elements"],
    template="""

        Your task is to analyse Entities, Value Objects in detail and generate the following :
        1. Generate the Entity Relationship for the Entities and Value Objects
        2. Give me diagram using mermaid as syntax in .md format 

        Service Context: {service_elements}
        
    """
)

service_er_chain = LLMChain(llm=model,prompt=service_class_diagram_prompt) 

final_chain = SequentialChain(chains=[service_er_chain],input_variables=["service_elements"],verbose=True)
response = final_chain.invoke({'service_elements': service_elements})

print(response['text'])
```

```python
service_openapi_prompt = PromptTemplate(
    input_variables=["service_elements"],
    template="""

    Your task is to analyse service in detail and generate the following :
        1. Generate the API operations using RESTful API. Make sure that POST,PUT,DELETE,PATCH are completed for each resource.
        2. Make sure that all the attribute and operations are using snake_case syntax.
        3. Give detailed descriptions, summary and title when description are allowed.
        4. Put examples in each operation and schema, you can use faker to generate random data. Follow the field specification.
        5. Each resource should have your own tag.
        6. Give me full openapi in the last version in yaml format.

    Service Context: {service_elements}
        
    """
)

service_er_chain = LLMChain(llm=model,prompt=service_openapi_prompt) 

final_chain = SequentialChain(chains=[service_er_chain],input_variables=["service_elements"],verbose=True)
response = final_chain.invoke({'service_elements': service_elements})

print(response['text'])
```
