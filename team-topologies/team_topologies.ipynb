```python
subdomains = """
    1. Fraud Detection and Prevention (Core Subdomain)
    2. Transaction Authorization (Core Subdomain)
    3. Payment Data Encryption (Supporting Subdomain)
    4. Compliance and Regulatory Management (Supporting Subdomain)
    5. Payment Method Integration (Supporting Subdomain)
    6. Merchant and Customer Management (Supporting Subdomain)
    7. Transaction Reporting and Analytics (Supporting Subdomain)
"""
```

```python
from langchain_openai import ChatOpenAI
import os
from langchain_core.messages import SystemMessage,AIMessage
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains import SequentialChain

model = ChatOpenAI(
    model="gpt-4o",
    temperature=0,
)

messages = [
    SystemMessage(content="You are a solution architect. Your main assignment is suggest team structure to enable fast flow deliveries.")
]

bounded_context_prompt = PromptTemplate(
    input_variables=["subdomains"],
    template="""

        Our company is aiming to structure our teams to develop a digital product efficiently. We have 30 developers and 4 Scrum Masters at our disposal, and we want to utilize the Team Topologies framework and Inverse Conway's Maneuver Law to guide our team structuring.
        
        Team Topologies Overview: We need an overview of how Team Topologies recommends structuring teams to optimize software delivery and DevOps practices. Include an explanation of the four fundamental team types (Stream-aligned Team, Enabling Team, Complicated Subsystem Team, and Platform Team) and the three team interaction modes (Collaboration, X-as-a-Service, Facilitating).
        
        Application of Conway’s Law: Explain Conway's Law, which states that organizations design systems that mirror their own communication structure. Discuss how this law can influence our team structure decisions, ensuring that our organizational design supports the architecture of the digital product we aim to develop.
        
        Proposed Team Structure: Based on the Team Topologies framework and an understanding of Conway's Law, propose a potential team structure for our scenario with 30 developers and 4 Scrum Masters. Consider how to divide these resources into the four types of teams, suggest the size of each team, and describe how these teams might interact to optimize workflow and product development.
        
        Role of Scrum Masters: Define the role of the Scrum Masters in this proposed structure. How would they facilitate the Agile processes, ensure effective communication between teams, and help in adhering to the best practices suggested by Team Topologies and Conway’s Law?
        
        Implementation Challenges: Identify potential challenges we might face in implementing this team structure and suggest strategies to overcome these challenges. Include considerations for ensuring smooth communication, maintaining team autonomy, and avoiding silos.
        
        Metrics for Success: Suggest metrics or indicators we should monitor to evaluate the effectiveness of the new team structure in fostering collaboration, speeding up delivery, and enhancing product quality.
        
        
        Your task is to perform the following actions: 
        
          1. Define fake names for each developer and scrumaster.
          2. The fake names for teams. 
          3. Assign each team for a group of subdomains.
          4. Scrumaster could be assigned to more than one team if it's needed. 
          5. We are in favor to use X-as-a-Service interaction mode for stream-aligned teams. It decouple a little bit our dependencies.
          6. We don't need any source code
        
        Use the following format:
          Subdomain: <Name> 
          Team: <Team Name>
        
        Subdomains: {subdomains}
        
    """
)

ddd_chain = LLMChain(llm=model,prompt=bounded_context_prompt) 

final_chain = SequentialChain(chains=[ddd_chain],input_variables=["subdomains"],verbose=True)

parsed_chain = final_chain

response = parsed_chain.invoke({'subdomains': subdomains})

teams_response = response['text']

print(response['text'])
```

```python

teams_prompt = PromptTemplate(
    input_variables=["teams_response"],
    template="""

        Give me a table in .md format with the subdomain, classification, team name, developers, scrumaster, team type and interaction modes
    
        Team Context: {teams_response}
        
    """
)

teams_chain = LLMChain(llm=model,prompt=teams_prompt) 

final_chain = SequentialChain(chains=[teams_chain],input_variables=["teams_response"],verbose=True)

parsed_chain = final_chain

response = parsed_chain.invoke({'teams_response': teams_response})

service_elements = response['text']

print(response['text'])
```
